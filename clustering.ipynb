```python
x=data.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
x=s.fit_transform(x)
```

```python
x
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++')  
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
```

```python
sns.regplot(data=data,x='Annual Income (k$)',y='Spending Score (1-100)',fit_reg=False)
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)
y_means=kmeans.fit_predict(x)
```

```python
y_means
```

```python
pd.concat([data,pd.DataFrame(y_means)],axis=1)
```

```python
plt.scatter(x[y_means==0,0],x[y_means==0,1],s=100,c='red',label='cluster1')
plt.scatter(x[y_means==1,0],x[y_means==1,1],s=100,c='pink',label='cluster2')
plt.scatter(x[y_means==2,0],x[y_means==2,1],s=100,c='blue',label='cluster3')
plt.scatter(x[y_means==3,0],x[y_means==3,1],s=100,c='green',label='cluster4')
plt.scatter(x[y_means==4,0],x[y_means==4,1],s=100,c='yellow',label='cluster5')
```

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('dendrogram')
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
```

```python
y_hc=hc.fit_predict(x)
```

```python
y_hc
```

```python
plt.scatter(x[y_hc==0,0],x[y_hc==0,1],s=100,c='red',label='cluster1')
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],s=100,c='green',label='cluster2')
plt.scatter(x[y_hc==2,0],x[y_hc==2,1],s=100,c='pink',label='cluster3')
plt.scatter(x[y_hc==3,0],x[y_hc==3,1],s=100,c='grey',label='cluster4')
plt.scatter(x[y_hc==4,0],x[y_hc==4,1],s=100,c='black',label='cluster5')
```

```python
pd.concat([data,pd.DataFrame(y_hc)],axis=1)
```

```python
from sklearn.metrics import silhouette_score
print(silhouette_score(x,y_means))
```
